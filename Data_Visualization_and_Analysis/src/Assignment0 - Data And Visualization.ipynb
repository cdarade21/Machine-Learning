##  Linear Algebra


  Understanding linear algebra is essential for understanding and working with machine learning algorithms like deep learning algorithms.It gives you better intution that how algorithms work under the hood. It's basically represented in the form of linear equations, and these equations are represented in the form of vectors and matrices.
  
  Linear algebra has very important significance in machine learning areas like vectorization, image recognition, dimensionality reduction, linear regression, regularization, pricipal component analysis and so many. It is helpful is reducing complexity of an algorithm/problem via it's inherent equations or properties. 
  
  In simple words,for computing multiple values simultaniously, vectorised code i.e using matrices and vectors is very effitient and reliable, saves computation time and developer's efforts. In image recognition, we can easily pass different images to the machine in it's understandable format i.e using matrix format, in order to perform wide variety of operations on it, for example classifying images. Many times problem seems complex due to wide number of it's features, which we can make simpler by reducing or combining it's features without loosing data significance by using linear algebra, which we call as 'Dimensionality Reduction', an essential part of 'Principal Component Analysis(PCA)'. Matrix decompositions are a useful tool for reducing a matrix to their constituent parts in order to simplify a range of more complex operations. The most used type of matrix decomposition is the 'eigendecomposition', in which we decompose a matrix into a set of eigenvectors and eigenvalues. Another popular dimensionality reduction method is the 'singular-value decomposition'(SVD). One of the most basic but important feature of linear algebra is reading data and datasets in the form of matrix (key structure of LA), so each row would be an obevation and each column would be a feature. In linear regression, numerical values can be predicted in simpler regression problems using linear algebra. Simpler models are often better at generalizing from specific examples to unseen data. In methods like neural networks, regression methods , these simple models are characterized by models having smaller coeffitient values. This technique to encourage a model to minimize the size of coeffitients while it is being fit on data is called 'regularization', achieved by linear algebra.
  

## Probability Theory

   Machine learning is bound to deal with uncertainities and nondeterministic quanities. It's difficult to say or think, that any event will be guranteedly occured unless its theoritically proven mathematically. So in general, these uncertainities can not be ignored and some framework is needed for quatifying these uncertanities, which is 'Probability Theory'. It allows us to be reasonable in situations where certainity is impossible. So, basically it provides some rules to determine prepositions as either True or False.
   
   To understand the probability in machine learning, few basic concepts are necessary to understand. So to start randomly through sample space, one 'random variable' is required which randomly takes on values from a sample space and, to understand its possible value, 'probability distribution' concept comes into picture. These probability distributions can either be described using 'PDF'(Probability density function) or 'PMF'(Probability Mass function).
Probability density function work on continuous variables, whereas, probability mass function works on discrete variables. Sometimes PMF work on multiple variables as well, such case is described as 'Joint probability Distribution'. To focus on particular subset of a set of random variables, 'Marginal Probability' is useful. It is found by using summation for discrete random variables and, integration for continuous random variables.  Other probabilty is 'Conditional Probability' of an event which depends upon some other event, and cannot be computed for an event that never happens. Decomposition of a Joint probability distribution into Conditional probability is also possible over any of it's random variable using 'Chain/Product Rule'.'Multinoulli/Categorical Distribution' is available on single discrete variable in a set of finite variables and both of these only for discrete variables. Common discrete distribution is the 'Bernoulli'and common distribution on real numbers is 'Gaussian/Normal Distribution'.

   Random variables also provides some functions like Expectation, Variance and Covariance, moments etc. 'Expectation', with respect to probability distribution is its mean value or average. 'Variance' gives a measure of how much the values of a function of a random variable x vary as we sample diﬀerent values of x from its probability distribution and 'Covariance' gives some sense of how much two values are linearly relatedto each other, as well as the scale of these variables.

## Introduction of data for Regression & Source

This dataset comprises prices of 50,000 round cut diamonds.
It's containing the prices and other attributes of almost 54,000 diamonds. The variables are as follows:

* File information : 

diamonds.csv (53940*10)

* Attribute Information: 

A data frame with 53940 rows and 10 variables:

1 : price
    price in US dollars (\$326–\$18,823)

2 : carat
    weight of the diamond (0.2–5.01)

3 : cut
    quality of the cut (Fair, Good, Very Good, Premium, Ideal)

4 : color
    diamond colour, from J (worst) to D (best)

5 : clarity
    a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

6 : x
    length in mm (0–10.74)

7 : y
    width in mm (0–58.9)

8 : z
    depth in mm (0–31.8)

9 : depth
    total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43–79)

10: table
    width of top of diamond relative to widest point (43–95)

Additional feature ‘dimension ’is added to understand the volume(i.e dimension) of diamond.

11: dimension
    volume of the diamond

* Source : https://vincentarelbundock.github.io/Rdatasets/datasets.html

## Reading the Data


Data file(diamonds.csv) is read with pandas dataframe.
It's shape and info is being displayed.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = pd.read_csv('diamonds.csv',sep=',',index_col=0)
diamonds_data.head()
```

```python
diamonds_data.shape
```

```python
diamonds_data.info()
```

## Preprocessing of the Data

In preprocessing activity, some column values('cut','color','clarity') are replaced with numbers as per level of precedence(priority) i.e 1 - Highest priority.

One new column is added which is used to determine the volume of a diamond, calculated with the product of length, width and depth of it.

```python
diamonds_data['cut'].replace({"Ideal": 1, 
                              "Premium":2,
                              "Very Good":3,
                              "Good":4,
                              "Fair":5},
                               inplace = True) 

diamonds_data['color'].replace({"D": 1, 
                              "E":2,
                              "F":3,
                              "G":4,
                              "H":5,
                              "I":6,
                              "J":7},
                               inplace = True)

diamonds_data['clarity'].replace({"IF": 1, 
                              "VVS1":2,
                              "VVS2":3,
                              "VS1":4,
                              "VS2":5,
                              "SI1":6,
                              "SI2":7,
                              "I1":8},
                               inplace = True)

diamonds_data['dimension'] = diamonds_data['x'] * diamonds_data['y'] * diamonds_data['z']
diamonds_data.head()
```

```python
diamonds_data.describe()
```

## Introduction of data for Classification & source

Two datasets are included, related to red and white vinho verde wine samples, from the north of Portugal. The goal is to model wine quality based on physicochemical tests.

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine.

* File information : 

winequality-red2.csv
winequality-white2.csv

(12994*14)

* Attribute Information:

For more information
Input variables (based on physicochemical tests):

1 : fixed acidity  
    Reducing acids significantly might lead to wines tasting flat. Fixed acids include tartaric, malic, citric, and       succinic acids)
    
2 : volatile acidity
    These acids are to be distilled out from the wine before completing the production process.
    
3 : citric acid
    This is one of the fixed acids which gives a wine its freshness.
    
4 : residual sugar
    This typically refers to the natural sugar from grapes which remains after the fermentation process stops, or is     stopped.
    
5 : chlorides
    This is usually a major contributor to saltiness in wine.
    
6 : free sulfur dioxide
    This is the part of the sulphur dioxide that when added to a wine is said to be free after the remaining part         binds
    
7 : total sulfur dioxide
    This is the sum total of the bound and the free sulfur dioxide
    This is mainly added to kill harmful bacteria and preserve quality and freshness. 
    
8 : density
    This can be represented as a comparison of the weight of a specific volume of wine to an equivalent volume of         water.It is generally used as a measure of the conversion of sugar to alcohol. 
    
9 : pH
    Also known as the potential of hydrogen, this is a numeric scale to specify the acidity or basicity the wine.
    
10: sulphates
    These are mineral salts containing sulfur. Sulphates are to wine as gluten is to food. They are a regular part of     the winemaking around the world and are considered essential.
    
11: alcohol
    Wine is an alcoholic beverage. Alcohol is formed as a result of yeast converting sugar during the fermentation       process. The percentage of alcohol can vary from wine to wine.
    
12: quality (score between 0 and 10)
    Wine experts graded the wine quality between 0 (very bad) and 10 (very excellent). The eventual quality score is     the median of at least three evaluations made by the same wine experts.
    
Additional two features are added :
    First is to determine type of wine and second is to determine wine quality.
    
13: wine_type
    Type of wine(red or white)
    
14: wine_quality
    Quality of the wine(low or mediun or high)
    
Source : https://archive.ics.uci.edu/ml/datasets/Wine+Quality

## Reading the Data

Both data files(winequality-red2.csv, winequality-white2.csv) are read with pandas dataframes and concatenated into
single one.

It's shape and info is being displayed after merging them in preprossessing section.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

red_wine_data = pd.read_csv('winequality-red2.csv',sep=',')
red_wine_data.head()
```

```python
white_wine_data = pd.read_csv('winequality-white2.csv',sep=',')
white_wine_data.head()
```

## Preprocessing of the Data

In preprocessing activity, Additional two features are added :
    First is to determine type of wine and second is to determine wine quality.

Then both datasets are concatenated into single dataframe.

```python
red_wine_data['wine_type'] = 'red'
red_wine_data['wine_quality'] = red_wine_data['quality'].apply(lambda value: 'low' 
                                                          if value <= 5 else 'medium' 
                                                              if value <= 7 else 'high')
red_wine_data['wine_quality'] = pd.Categorical(red_wine_data['wine_quality'], 
                                           categories=['low', 'medium', 'high'])
red_wine_data.head()
```

```python
white_wine_data['wine_type'] = 'white'
white_wine_data['wine_quality'] = white_wine_data['quality'].apply(lambda value: 'low' 
                                                              if value <= 5 else 'medium' 
                                                                  if value <= 7 else 'high')
white_wine_data['wine_quality'] = pd.Categorical(white_wine_data['wine_quality'], 
                                             categories=['low', 'medium', 'high'])
white_wine_data.head()
```

```python
wine_data=pd.concat([red_wine_data,white_wine_data])
wine_data = wine_data.sample(frac=1, random_state=95).reset_index(drop=True)
wine_data.head()
```

```python
wine_data.shape
```

```python
wine_data.info()
```

## Visualization of the Data and Preliminary obervations

```python
data_cols = ['residual sugar', 'total sulfur dioxide', 'sulphates', 'alcohol', 'volatile acidity', 'quality']
rs = round(red_wine_data[data_cols].describe(),2)
ws = round(white_wine_data[data_cols].describe(),2)
pd.concat([rs, ws], axis=1, keys=['Red Wine Stats', 'White Wine Stats'])
```

```python
data_cols = ['alcohol', 'volatile acidity', 'pH', 'quality']
ls = round(wine_data[wine_data['wine_quality'] == 'low'][data_cols].describe(),2)
ms = round(wine_data[wine_data['wine_quality'] == 'medium'][data_cols].describe(),2)
hs = round(wine_data[wine_data['wine_quality'] == 'high'][data_cols].describe(),2)
pd.concat([ls, ms, hs], axis=1, keys=['Low Quality Wine', 'Medium Quality Wine', 'High Quality Wine'])
```

```python
wines_corr = wine_data.corr()
round(wines_corr,2)
```

```python
sns.heatmap(round(wines_corr,2),cmap='coolwarm')
plt.show()
```
